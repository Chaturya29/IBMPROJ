```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read the dataset
data = pd.read_excel('/content/DoctorVisits (2).xlsx')
```

```python
print("Dimensions: ", data.shape)
```

```python
print(data.head())
```

```python
print(data.dtypes)
```

```python
print(data.isnull().sum())
```

```python
data = data.fillna(data.mean())
```

```python
plt.scatter(data['age'], data['visits'])
plt.xlabel('Age')
plt.ylabel('Visits')
plt.title('Age vs. Visits')
plt.show()
```

```python
gender_counts = data['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.xlabel('Gender')
plt.ylabel('Visits')
plt.title('Gender vs. Visits')
plt.show()
```

```python
age_avg_income = data.groupby('age')['income'].mean()
plt.plot(age_avg_income.index, age_avg_income.values)
plt.xlabel('Age')
plt.ylabel('Average Income')
plt.title('Age vs. Average Income')
plt.show()
```
